```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_drug_df, clinical_trial_df, on="Mouse ID")

# Display the data table for preview
combined_data.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_group = combined_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)']
mean_drug=drug_group.mean()

# Convert to DataFrame
drug_tumor_df=mean_drug.reset_index()


# Preview DataFrame
drug_tumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_error = drug_group.sem()

# Convert to DataFrame
tumor_volume_error_df = pd.DataFrame(tumor_volume_error)

# Preview DataFrame
tumor_volume_error_df.reset_index(inplace=True)
tumor_volume_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_pivot= drug_tumor_df.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")

#drug_pivot.reset_index(level=None, inplace=True)

# Preview that Reformatting worked
drug_pivot.head(10)
```

```python
# Generate the Plot (with Error Bars)
drug_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]

for drug,maker,color in drug_list:
    plot_drug = plt.errorbar(drug_pivot.index,drug_pivot[drug],fmt='^',ls='--',linewidth=0.5)

    
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
# Save the Figure

```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_gb = combined_data.groupby(['Drug','Timepoint'])

# Convert to DataFrame
meta_df = pd.DataFrame(meta_gb['Metastatic Sites'].mean())

# Preview DataFrame
meta_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_pivot=meta_df.pivot_table(index="Timepoint", columns="Drug")["Metastatic Sites"]


# Preview that Reformatting worked
meta_pivot.head(10)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_error= meta_gb['Metastatic Sites'].sem()

# Convert to DataFrame
#meta_df.reset_index(inplace=True)
meta_sem=pd.DataFrame(meta_error)

# Preview DataFrame
meta_sem.reset_index(inplace=True)
meta_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_pivot_sem=meta_sem.pivot_table(index="Timepoint", columns="Drug")["Metastatic Sites"]


# Preview that Reformatting worked
meta_pivot_sem.head(10)
```

```python
# Generate the Plot (with Error Bars)
drug_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]

for drug,maker,color in drug_list:
    plot_meta = plt.errorbar(meta_pivot.index,meta_pivot[drug],fmt='^',ls='--',linewidth=0.5)

    
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic")
plt.grid()

# Save the Figure

# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_gb = combined_data.groupby(['Drug','Timepoint']).count()['Tumor Volume (mm3)']

# Convert to DataFrame
survival_df = pd.DataFrame({'Mouse Count':survival_gb})

# Preview DataFrame
survival_df.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_pivot=survival_df.pivot_table(index="Timepoint", columns="Drug")["Mouse Count"]

# Preview the Data Frame
mouse_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
drug_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]

for drug,maker,color in drug_list:
    plot_mouse = plt.plot(mouse_pivot.index,mouse_pivot[drug],linewidth=2)

    
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()
# Save the Figure

# Show the Figure
plt.show()
```
